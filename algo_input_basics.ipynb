## 9. 常见错误与修正

最常见的 4 个错误：

1. 忘记 `int()` 转换
2. `split()` 后没转整数
3. 循环次数写错
4. 没看清输入格式

````python
# 错误 1：忘记转 int
s = "123"
print(s + "1")
print(int(s) + 1)

# 错误 2：split() 后没转整数
parts = input().split()
print(parts)
print(list(map(int, parts)))

# 错误 3：循环次数写错
n = int(input())
for _ in range(n):
    x = int(input())
    print(x)
````

# 杨辉三角

````python
n = int(input())


nums = [[1] * n for _ in range(n)] # 生成一个 n x n 的二维列表，初始值都是 1
# 其中 [1] * n 生成一个长度为 n 的列表，元素都是 1；
# 外层的 for _ in range(n) 生成了 n 个这样的列表，形成了一个 n x n 的二维列表。

for i in range(n): 
    for j in range(1, i): # 注意 j 的范围，是因为边界上的元素都是 1
        nums[i][j] = nums[i - 1][j - 1] + nums[i - 1][j]


for i in range(n):
    for j in range(i + 1): # 注意 j 的范围，打印到 [0, i] 就行了，因为后面都是 1
        print(nums[i][j], end = ' ')
    print()
````

## 10. 队列 deque 基础 API

算法竞赛里，队列通常用 `collections.deque`，因为左侧弹出是 `O(1)`。

样例输入：

```text
5
1 2 3 4 5
```

````python
from collections import deque # 导入库，deque内置库

n = int(input())
nums = list(map(int, input().split()))

q = deque(nums) # 用现有列表直接建队列

q.append(6) # 队尾加入一个元素
left = q.popleft() # 队头弹出一个元素，复杂度 O(1)

print(left) # 1
print(q[0]) # 当前队头
print(len(q)) # 当前队列长度
print(list(q))
````

## 11. 栈 list 基础 API

算法竞赛里，栈通常直接用 `list` 模拟，常见操作就是 `append()` 和 `pop()`。

样例输入：

```text
5
1 2 3 4 5
```

````python
n = int(input())
nums = list(map(int, input().split())) # [1, 2, 3, 4, 5]

stack = [] # 表示一个栈，初始为空

for x in nums:
    stack.append(x) # 入栈，放到末尾


print(stack[-1]) # 取栈顶元素
top = stack.pop() # 出栈，弹出末尾元素


print(top) # 5
print(len(stack)) # 当前栈长度
print(stack)
````

## 13. 字典基本操作 API

字典在算法竞赛里最常见的用途是做“计数”和“快速查询”。

样例输入：

```text
5
1 2 2 3 1
2
```

````python
n = int(input())
nums = list(map(int, input().split()))
x = int(input())

cnt = {}

for num in nums:
    cnt[num] = cnt.get(num, 0) + 1 # 统计每个数字出现次数

print(cnt.get(x, 0)) # 查询 x 出现了多少次，不存在就返回 0
print(x in cnt) # 判断 x 是否出现过

cnt[4] = 7 # 新增或修改一个键值对
del cnt[4] # 删除一个键


print(list(cnt.keys())) # 所有键
print(cnt.values()) # 所有值
````

## 14. 一维列表 nums 的遍历与打印

一维列表最常见的输出方式有 5 类：

1. 直接整体打印
2. 逐个元素打印
3. 带下标遍历打印
4. 按一行格式化输出
5. 用 `end=' '` 控制同一行输出

样例输入：

```text
5
1 2 3 4 5
```

下面代码里的注释，会直接写出常见输出效果。

````python
n = int(input())
nums = list(map(int, input().split()))

print(nums) # 输出: [1, 2, 3, 4, 5]

# 逐个元素打印，每个数字占一行
# 输出:
# 1
# 2
# 3
# 4
# 5
for num in nums:
    print(num)

# 打印下标和元素
# 输出:
# 0 1
# 1 2
# 2 3
# 3 4
# 4 5
for i in range(n):
    print(i, nums[i])

# enumerate 写法更常用，效果和上面类似
# 输出:
# 0 1
# 1 2
# 2 3
# 3 4
# 4 5
for i, num in enumerate(nums):
    print(i, num)

print(*nums) # 输出: 1 2 3 4 5
print(' '.join(map(str, nums))) # 输出: 1 2 3 4 5

# 这也是算法题里很常见的一种写法
# 输出: 1 2 3 4 5
for num in nums:
    print(num, end=' ')
print()
````

## 15. 二维数组 grid 的遍历与打印

二维数组常见的遍历方式有：

1. 直接打印整个二维列表
2. 按行打印
3. 每行按 `1 2 3` 这种格式输出
4. 逐个元素遍历
5. 带行列坐标遍历
6. 用 `end=' '` 控制矩阵输出格式

样例输入：

```text
2 3
1 2 3
4 5 6
```

下面代码里的注释，会直接写出常见输出效果。

````python
n, m = map(int, input().split())
grid = [list(map(int, input().split())) for _ in range(n)]

print(grid) # 输出: [[1, 2, 3], [4, 5, 6]]

# 每次打印一整行
# 输出:
# [1, 2, 3]
# [4, 5, 6]
for row in grid:
    print(row)

# 每行按题目常见格式输出
# 输出:
# 1 2 3
# 4 5 6
for row in grid:
    print(*row)

# 逐个元素打印，每个数字占一行
# 输出:
# 1
# 2
# 3
# 4
# 5
# 6
for i in range(n):
    for j in range(m):
        print(grid[i][j])

# 打印行号、列号和元素值
# 输出:
# 0 0 1
# 0 1 2
# 0 2 3
# 1 0 4
# 1 1 5
# 1 2 6
for i in range(n):
    for j in range(m):
        print(i, j, grid[i][j])

# 用 end=' ' 手动控制每一行的输出格式
# 输出:
# 1 2 3
# 4 5 6
for row in grid:
    for num in row:
        print(num, end=' ')
    print()
````
